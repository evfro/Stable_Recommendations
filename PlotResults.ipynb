```python
import matplotlib.pyplot as plt
import seaborn as sns
from datareader import read_data, algs, display_ranks, combine_all_metrics

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
sns.set("paper", style="ticks", font_scale=2.5)
```

```python
plot_params = dict(
    x='rank', y='value', kind='box', col='metric', hue='model', sharey=False, palette='vlag',
    aspect=1.5,
    height=5,
)
```

# Movielens- New (with filtered Ratings)

```python
data, scores, test_users = read_data('Results for ML_new', 'ML')
ml_metrics = (combine_all_metrics(scores, data)
    .query('rank in @display_ranks')
    .set_index(['rank', 'step', 'model'])
    .drop(['HR'], axis=1)
    .stack()
    .to_frame('value')
    .reset_index()
    .rename(columns={'level_3': 'metric'})
)
ml_metrics.head()
```

```python
g = sns.catplot(data=ml_metrics, **plot_params)   ##New ML
g.set(xlabel='')
plt.savefig('ml_results.pdf', dpi=400)
```
